#### Removing Outliers on Standard Deviation Distance From Mean

```python
# Find the mean of the height-set
mean = weight_height_dataframe.Height.mean()
```

```python
# Then find the standard deviation of the height-set
std_deviation = weight_height_dataframe.Height.std()
```

```python
# Let's assume we want to rule-out as outliers any data-points that are more than 3 standard deviations away from the mean.
# Lit. 3-sig that shi.
mid_pass_lower_bound = mean - (3 * std_deviation)
upper_pass_upper_bound = mean + (3 * std_deviation)
print("Lower bound: ", mid_pass_lower_bound)
print("Upper bound: ", upper_pass_upper_bound)
```
